```python
from collections import deque

def print_state(state):
    print(state[0] + "|" + state[1] + "|" + state[2] + "|")
    print("------")
    print(state[3] + "|" + state[4] + "|" + state[5] + "|" )
    print("------")
    print(state[6] + "|" + state[7] + "|" + state[8] + "|")
    print("\n")

def check_win(state):
    #for row
    for x in range(0,9,3):
        if state[x]==state[x+1]==state[x+2] and state[x]!=" ":
            return True

    #for col
    for x in range(3):
        if state[x]==state[x+3]==state[x+6] and state[x]!=" ":
            return True

    #for diagonal

    if state[0]==state[4]==state[8] and state[0]!=" " :
        return True
    if state[2]==state[4]==state[6] and state[2]!=" ":
        return True

    return False

def get_empty_cells(state):
    """Return a list of empty cells in the board"""
    return [i for i in range(len(state)) if state[i] == " "]

def get_successors(state, player):
    """Return a list of possible successor states for the given player"""
    empty_cells = get_empty_cells(state)
    successors = []
    for cell in empty_cells:
        new_state = state[:cell] + player + state[cell+1:]
        successors.append(new_state)
    return successors

def dfs_search(initial_state):
    """Perform a DFS search to find a winning state and its parent states"""
    stack = []
    visited = set()
    parents = {}
    stack.append(initial_state)
    visited.add(initial_state)
    while stack:
        state = stack.pop()
        if check_win(state):
            winning_state = state
            break
        successors = get_successors(state, "X" if state.count("X") == state.count("O") else "O")
        for successor in successors:
            if successor not in visited:
                stack.append(successor)
                visited.add(successor)
                parents[successor] = state
    else:
        print("No winning state found.")
        return None, None
    print("Winning state:\n")
    print_state(winning_state)
    print("Parent states:\n")
    parent_states = [winning_state]
    parent_state = parents[winning_state]
    while parent_state != initial_state:
        parent_states.append(parent_state)
        parent_state = parents[parent_state]
    parent_states.append(initial_state)
    for state in reversed(parent_states):
        print_state(state)
    return winning_state, parents

def play_game():
    initial_state = "         "
    print_state(initial_state)
    winning_state, parents = dfs_search(initial_state)

if __name__ == "__main__":
    play_game()
```
